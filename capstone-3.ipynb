```python

df['Date'] = pd.to_datetime(
    
    df['Date']
)

print("Date Conversion Completed")
```

```python
df = df.sort_values(
    
    by='Date'
)

df = df.reset_index(
    
    drop=True
)

print("Dataset Sorted Chronologically")
```

```python
price_columns = [
    
    'Price',
    
    'Open',
    
    'High',
    
    'Low'
]

for col in price_columns:
    
    df[col] = df[col].str.replace(
        
        ',',
        
        ''
    )
    
    df[col] = df[col].astype(float)

print("Price Columns Cleaned Successfully")
```

```python

df['Change %'] = df['Change %'].str.replace(
    
    '%',
    
    ''
)

df['Change %'] = df['Change %'].astype(float)

print("Change Percentage Column Cleaned")
```

```python


def convert_volume(volume):
    
    if pd.isnull(volume):
        
        return np.nan
    
    volume = str(volume)
    
    if 'K' in volume:
        
        return float(
            
            volume.replace('K', '')
        ) * 1000
    
    elif 'M' in volume:
        
        return float(
            
            volume.replace('M', '')
        ) * 1000000
    
    else:
        
        try:
            
            return float(volume)
        
        except:
            
            return np.nan

df['Vol.'] = df['Vol.'].apply(
    
    convert_volume
)

print("Volume Column Cleaned Successfully")
```

```python
df['Vol.'] = df['Vol.'].fillna(
    
    df['Vol.'].median()
)

print("Missing Values Handled Successfully")
```

```python

df = df.drop_duplicates()

df = df.reset_index(
    
    drop=True
)

print("Duplicate Records Removed")
```

```python

df.set_index(
    
    'Date',
    
    inplace=True
)

print("Date Column Set as Index")
```

```python

df['Lag_1'] = df['Price'].shift(1)

df['Lag_7'] = df['Price'].shift(7)

df['Lag_30'] = df['Price'].shift(30)

print("Lag Features Created Successfully")
```

```python

df['Rolling_Mean_7'] = df['Price'].rolling(
    
    window=7
).mean()

df['Rolling_Mean_30'] = df['Price'].rolling(
    
    window=30
).mean()

df['Rolling_Mean_90'] = df['Price'].rolling(
    
    window=90
).mean()

print("Rolling Mean Features Created")
```

```python
df['Rolling_STD_7'] = df['Price'].rolling(
    
    window=7
).std()

df['Rolling_STD_30'] = df['Price'].rolling(
    
    window=30
).std()

print("Rolling Volatility Features Created")
```

```python

df['EMA_7'] = df['Price'].ewm(
    
    span=7,
    
    adjust=False
).mean()

df['EMA_30'] = df['Price'].ewm(
    
    span=30,
    
    adjust=False
).mean()

print("Exponential Moving Averages Created")
```

```python

df['Daily_Range'] = (
    
    df['High'] - df['Low']
)

print("Daily Range Feature Created")
```

```python

df['Momentum_7'] = (
    
    df['Price'] - df['Lag_7']
)

df['Momentum_30'] = (
    
    df['Price'] - df['Lag_30']
)

print("Momentum Features Created")
```

```python
df['ROC_7'] = (
    
    df['Price'].pct_change(periods=7)
)

df['ROC_30'] = (
    
    df['Price'].pct_change(periods=30)
)

print("Rate of Change Features Created")
```

```python
df['Volatility_30'] = (
    
    df['Rolling_STD_30'] / df['Rolling_Mean_30']
)

print("Volatility Indicator Created")
```

```python
df['Year'] = df.index.year

df['Month'] = df.index.month

df['Day'] = df.index.day

df['DayOfWeek'] = df.index.dayofweek

df['Quarter'] = df.index.quarter

print("Date-Based Features Created")
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.ensemble import RandomForestRegressor

from sklearn.ensemble import GradientBoostingRegressor

print("Regression Models Imported Successfully")
```

```python
from sklearn.metrics import (
    
    mean_absolute_error,
    
    mean_squared_error,
    
    r2_score
)

print("Evaluation Metrics Imported Successfully")
```

```python

models = {
    
    'Linear Regression' : LinearRegression(),
    
    'Random Forest Regressor' : RandomForestRegressor(
        
        random_state=42
    ),
    
    'Gradient Boosting Regressor' : GradientBoostingRegressor(
        
        random_state=42
    )
}

print("Models Initialized Successfully")
```

```python
model_results = []
```

```python


for model_name, model in models.items():
    
    print(f"\nTraining {model_name}...\n")
    
    # Training Model
    model.fit(
        
        X_train_scaled,
        
        y_train
    )
    
    # Predictions
    predictions = model.predict(
        
        X_test_scaled
    )
    
    # Evaluation Metrics
    mae = mean_absolute_error(
        
        y_test,
        
        predictions
    )
    
    mse = mean_squared_error(
        
        y_test,
        
        predictions
    )
    
    rmse = np.sqrt(mse)
    
    r2 = r2_score(
        
        y_test,
        
        predictions
    )
    
    # MAPE
    mape = np.mean(
        
        np.abs(
            
            (y_test - predictions) / y_test
        )
    ) * 100
    
    # Storing Results
    model_results.append({
        
        'Model' : model_name,
        
        'MAE' : mae,
        
        'RMSE' : rmse,
        
        'R2 Score' : r2,
        
        'MAPE' : mape
    })
    
    print(f"{model_name} Training Completed")
```

```python

results_df = pd.DataFrame(
    
    model_results
)

results_df
```

```python

best_model_name = results_df.sort_values(
    
    by='R2 Score',
    
    ascending=False
).iloc[0]['Model']

print("Best Forecasting Model :", best_model_name)
```

```python
best_model = models[best_model_name]

best_model.fit(
    
    X_train_scaled,
    
    y_train
)

print("Best Model Trained Successfully")
```

```python
# ============================================================
# COMPLETE FINAL MODEL COMPARISON PIPELINE
# INCLUDING TUNED MODEL + OVERFITTING ANALYSIS
# ============================================================

from sklearn.linear_model import LinearRegression

from sklearn.tree import DecisionTreeRegressor

from sklearn.ensemble import (
    
    RandomForestRegressor,
    
    GradientBoostingRegressor,
    
    ExtraTreesRegressor,
    
    AdaBoostRegressor
)

from sklearn.neighbors import KNeighborsRegressor

from sklearn.svm import SVR

from xgboost import XGBRegressor

from sklearn.model_selection import RandomizedSearchCV

from sklearn.metrics import (
    
    mean_absolute_error,
    
    mean_squared_error,
    
    r2_score
)

# ============================================================
# BASE MODELS
# ============================================================

models = {
    
    'Linear Regression' : LinearRegression(),
    
    'Decision Tree' : DecisionTreeRegressor(
        
        random_state=42
    ),
    
    'Random Forest' : RandomForestRegressor(
        
        n_estimators=100,
        
        random_state=42
    ),
    
    'Gradient Boosting' : GradientBoostingRegressor(
        
        random_state=42
    ),
    
    'Extra Trees' : ExtraTreesRegressor(
        
        n_estimators=100,
        
        random_state=42
    ),
    
    'AdaBoost' : AdaBoostRegressor(
        
        random_state=42
    ),
    
    'KNN Regressor' : KNeighborsRegressor(),
    
    'SVR' : SVR(),
    
    'XGBoost' : XGBRegressor(
        
        random_state=42,
        
        n_estimators=100
    )
}

# ============================================================
# RESULTS STORAGE
# ============================================================

all_model_results = []

# ============================================================
# TRAINING BASE MODELS
# ============================================================

for model_name, model in models.items():
    
    print(f"\n==============================")
    
    print(f"TRAINING : {model_name}")
    
    print(f"==============================")
    
    # ========================================================
    # TRAINING
    # ========================================================
    
    model.fit(
        
        X_train_scaled,
        
        y_train
    )
    
    # ========================================================
    # PREDICTIONS
    # ========================================================
    
    train_pred = model.predict(
        
        X_train_scaled
    )
    
    test_pred = model.predict(
        
        X_test_scaled
    )
    
    # ========================================================
    # METRICS
    # ========================================================
    
    train_r2 = r2_score(
        
        y_train,
        
        train_pred
    )
    
    test_r2 = r2_score(
        
        y_test,
        
        test_pred
    )
    
    mae = mean_absolute_error(
        
        y_test,
        
        test_pred
    )
    
    rmse = np.sqrt(
        
        mean_squared_error(
            
            y_test,
            
            test_pred
        )
    )
    
    mape = np.mean(
        
        np.abs(
            
            (y_test - test_pred) / y_test
        )
    ) * 100
    
    overfit_gap = abs(
        
        train_r2 - test_r2
    )
    
    # ========================================================
    # STORING RESULTS
    # ========================================================
    
    all_model_results.append({
        
        'Model' : model_name,
        
        'Train R2' : round(train_r2, 4),
        
        'Test R2' : round(test_r2, 4),
        
        'MAE' : round(mae, 4),
        
        'RMSE' : round(rmse, 4),
        
        'MAPE' : round(mape, 4),
        
        'Overfitting Gap' : round(overfit_gap, 4)
    })

# ============================================================
# HYPERPARAMETER TUNING
# ============================================================

print("\n==============================")

print("TRAINING : OPTIMIZED RANDOM FOREST")

print("==============================")

rf_model = RandomForestRegressor(
    
    random_state=42
)

param_grid = {
    
    'n_estimators' : [50, 100],
    
    'max_depth' : [5, 10],
    
    'min_samples_split' : [2, 5],
    
    'min_samples_leaf' : [1, 2]
}

random_search = RandomizedSearchCV(
    
    estimator=rf_model,
    
    param_distributions=param_grid,
    
    n_iter=5,
    
    scoring='r2',
    
    cv=3,
    
    random_state=42,
    
    verbose=1,
    
    n_jobs=-1
)

random_search.fit(
    
    X_train_scaled,
    
    y_train
)

optimized_model = random_search.best_estimator_

# ============================================================
# OPTIMIZED MODEL PREDICTIONS
# ============================================================

optimized_train_pred = optimized_model.predict(
    
    X_train_scaled
)

optimized_test_pred = optimized_model.predict(
    
    X_test_scaled
)

# ============================================================
# OPTIMIZED METRICS
# ============================================================

optimized_train_r2 = r2_score(
    
    y_train,
    
    optimized_train_pred
)

optimized_test_r2 = r2_score(
    
    y_test,
    
    optimized_test_pred
)

optimized_mae = mean_absolute_error(
    
    y_test,
    
    optimized_test_pred
)

optimized_rmse = np.sqrt(
    
    mean_squared_error(
        
        y_test,
        
        optimized_test_pred
    )
)

optimized_mape = np.mean(
    
    np.abs(
        
        (y_test - optimized_test_pred) / y_test
    )
) * 100

optimized_gap = abs(
    
    optimized_train_r2 - optimized_test_r2
)

# ============================================================
# STORING OPTIMIZED MODEL RESULTS
# ============================================================

all_model_results.append({
    
    'Model' : 'Optimized Random Forest',
    
    'Train R2' : round(optimized_train_r2, 4),
    
    'Test R2' : round(optimized_test_r2, 4),
    
    'MAE' : round(optimized_mae, 4),
    
    'RMSE' : round(optimized_rmse, 4),
    
    'MAPE' : round(optimized_mape, 4),
    
    'Overfitting Gap' : round(optimized_gap, 4)
})

# ============================================================
# FINAL RESULTS TABLE
# ============================================================

all_results_df = pd.DataFrame(
    
    all_model_results
)

all_results_df = all_results_df.sort_values(
    
    by='Test R2',
    
    ascending=False
)

print("\n=================================================")

print("FINAL MODEL COMPARISON")

print("=================================================\n")

display(all_results_df)

# ============================================================
# VISUALIZING TEST R2 SCORES
# ============================================================

plt.figure(figsize=(14,6))

sns.barplot(
    
    data=all_results_df,
    
    x='Model',
    
    y='Test R2'
)

plt.xticks(rotation=45)

plt.title('Forecasting Model Test R2 Comparison')

plt.grid()

plt.show()

# ============================================================
# VISUALIZING OVERFITTING GAP
# ============================================================

plt.figure(figsize=(14,6))

sns.barplot(
    
    data=all_results_df,
    
    x='Model',
    
    y='Overfitting Gap'
)

plt.xticks(rotation=45)

plt.title('Model Overfitting Gap Comparison')

plt.grid()

plt.show()
```
